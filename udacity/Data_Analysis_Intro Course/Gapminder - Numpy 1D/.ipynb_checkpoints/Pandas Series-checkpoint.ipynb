### Vectorized Operation on Series

```python
# Addition when indexes are the same
s1 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
print(s1 + s2)
```

```python
s1.loc['a'] 
```

```python
# Indexes have same elements in a different order
s1 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([10, 20, 30, 40], index=['b', 'd', 'a', 'c'])
print(s1 + s2)
```

```python
# Indexes overlap, but do not have exactly the same elements
s1 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([10, 20, 30, 40], index=['c', 'd', 'e', 'f'])
print(s1 + s2)
```

```python
# Indexes do not overlap
s1 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([10, 20, 30, 40], index=['e', 'f', 'g', 'h'])
print(s1 + s2)
```

```python
# dropna function on series
s1 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([10, 20, 30, 40], index=['c', 'd', 'e', 'f'])
sum_result = s1 + s2
print(sum_result.dropna())
```

```python
# fillna(0) function on series
s1 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([10, 20, 30, 40], index=['c', 'd', 'e', 'f'])
sum_result = s1 + s2
print(sum_result.fillna(0))
```

```python
# Filling missing value Treat missing value as 0 before addition
s1 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([10, 20, 30, 40], index=['c', 'd', 'e', 'f'])
sum_result = s1.add(s2, fill_value = 0)
print(sum_result)
```

### Pandas Series apply() function

```python
# apply() takes a Series and a function and returns a new Series
def add3(ser):
    return ser+3

a = pd.Series([1,2,3,4])
a.apply(add3)
```

### Plotting in Pandas

If the variable data is a NumPy array or a Pandas Series, just like if it is a list, the code

#### import matplotlib.pyplot as plt
#### plt.hist(data) will create a histogram of the data.

Pandas also has built-in plotting that uses matplotlib behind the scenes, so if data is a Series, you can create a histogram using 
#### data.hist().

There's no difference between these two in this case, but sometimes the Pandas wrapper can be more convenient. 
For example, you can make a line Plot of a series using data.plot()

The index of the Series will be used for the x-axis and the values for the y-axis.

```python
life_expectancy.plot();
```
